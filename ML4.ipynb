```python
import pandas as pd
```

```python
df = pd.read_csv('Churn_Modelling.csv')
df
```

```python
df.drop(columns=['RowNumber','CustomerId','Surname'],inplace=True)
```

```python
df
```

```python
geography = pd.get_dummies(df['Geography'])
gender = pd.get_dummies(df['Geography'])
```

```python
df = pd.concat([df,geography,gender],axis=1)
```

```python
df
```

```python
df.drop(columns=['Geography','Gender'],inplace=True)
```

```python
df
```

```python
x = df.drop(['Exited'],axis=1)
```

```python
y=df['Exited']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
x_train = scalar.fit_transform(x_train)
x_test = scalar.fit_transform(x_test)
```

```python
import keras 
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier = Sequential()
classifier.add(Dense(activation = 'relu',input_dim=14,units=6,kernel_initializer = 'uniform'))
classifier.add(Dense(activation = 'relu',units=6,kernel_initializer='uniform'))
classifier.add(Dense(activation = 'sigmoid',units=1))
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
classifier.summary()
```

```python
classifier.fit(x_train,y_train,epochs=10,batch_size=10)
```

```python
pred = classifier.predict(x_test)
```

```python
y_pred = pred>0.5
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
recall = recall_score(y_test,y_pred)
recall
```

```python
acc = accuracy_score(y_test,y_pred)
acc
```

```python
pre = precision_score(y_test,y_pred)
pre
```
