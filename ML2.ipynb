```python
import pandas as pd
```

```python
df = pd.read_csv('sales_data_sample.csv',encoding='latin1')
df
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df = df.drop(columns=['ADDRESSLINE1','ADDRESSLINE2','STATE','TERRITORY','ORDERNUMBER','POSTALCODE','PHONE','ORDERDATE','STATUS','QTR_ID','CUSTOMERNAME', 'CITY',
       'CONTACTLASTNAME', 'CONTACTFIRSTNAME'],axis=1)
```

```python
DEALSIZE = pd.get_dummies(df['DEALSIZE'])
```

```python
PRODUCTLINE = pd.get_dummies(df['PRODUCTLINE'])
```

```python
COUNTRY = pd.get_dummies(df['COUNTRY'])
```

```python
df = pd.concat([df,DEALSIZE,PRODUCTLINE,COUNTRY],axis=1)
df
```

```python
if 'DEALSIZE' in df and 'PRODUCTLINE' in df and 'COUNTRY' in df :
    df.drop(columns=['DEALSIZE','PRODUCTLINE','COUNTRY'],axis=1,inplace=True)
```

```python
df
```

```python
df['PRODUCTCODE']=pd.Categorical(df['PRODUCTCODE']).codes
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.columns
```

```python
from sklearn.cluster import KMeans
```

```python
distortion=[]
k=range(1,10)
for n in k:
    km=KMeans(n_clusters=n)
    km.fit(df)
    distortion.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(k,distortion,'bx-')
plt.xlabel('k')
plt.ylabel('distortion')
plt.show()
```

```python
x_train = df.values
model = KMeans(n_clusters=4,random_state=42)
model.fit(x_train)
```

```python
pred = model.predict(x_train)
```

```python
print(pred)
```

```python
pred1 = pd.DataFrame(pred)
```

```python
pred1
```

```python
df = pd.concat([df,pred1],axis=1)
```
